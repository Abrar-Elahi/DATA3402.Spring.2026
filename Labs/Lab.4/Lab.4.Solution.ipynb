1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter. 

```python
class Counter:
    def __init__(self, max_val):
        self.max_val = max_val
        self.current_val = 0
        
    def increment(self):
        if self.current_val < self.max_val:
            self.current_val += 1
            print(f"Current Count: {self.current_val}")
        else:
            print(f"Cannot add to count, max value of {self.max_val} reached")
            
    def reset(self):
        self.current_val = 0
        print("Counter reset to 0.")

    def get_count(self):
        return self.current_val
```

```python
class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        # Empty canvas is a matrix with element being the "space" character
        self.data = [[' '] * width for i in range(height)]

    def set_pixel(self, row, col, char='*'):
        self.data[row][col] = char

    def get_pixel(self, row, col):
        return self.data[row][col]
    
    def clear_canvas(self):
        self.data = [[' '] * self.width for i in range(self.height)]
    
    def v_line(self, x, y, w, **kargs):
        for i in range(x,x+w):
            self.set_pixel(i,y, **kargs)

    def h_line(self, x, y, h, **kargs):
        for i in range(y,y+h):
            self.set_pixel(x,i, **kargs)
            
    def line(self, x1, y1, x2, y2, **kargs):
        slope = (y2-y1) / (x2-x1)
        for y in range(y1,y2):
            x= int(slope * y)
            self.set_pixel(x,y, **kargs)
            
    def display(self):
        print("\n".join(["".join(row) for row in self.data]))
```

```python
#Had to flip triangle, so redid class

class Triangle(Shape):
    def __init__(self, length1, length2, base, height, x, y):
        super().__init__(x, y)
        self.__length1 = length1
        self.__length2 = length2
        self.__base = base
        self.__height = height

    def get_length1(self): return self.__length1
    def get_length2(self): return self.__length2
    def get_base(self): return self.__base
    def get_height(self): return self.__height
    def get_area(self): return 0.5 * (self.__base * self.__height)
    def get_perimeter(self): return self.__length1 + self.__length2 + self.__base

    def get_points_of_perimeter(self):
        x, y, b, h = self.get_x(), self.get_y(), self.__base, self.__height
        return [(x, y + h), (x + b, y + h), (x + (b / 2), y)]
    
    def is_inside(self, px, py):
        x1, y1 = self.get_x(), self.get_y() + self.__height
        x2, y2 = self.get_x() + self.__base, self.get_y() + self.__height
        x3, y3 = self.get_x() + (self.__base / 2), self.get_y()

        def calc_area(ax, ay, bx, by, cx, cy):
            return abs((ax * (by - cy) + bx * (cy - ay) + cx * (ay - by)) / 2.0)

        main_area = calc_area(x1, y1, x2, y2, x3, y3)
        area1 = calc_area(px, py, x2, y2, x3, y3)
        area2 = calc_area(x1, y1, px, py, x3, y3)
        area3 = calc_area(x1, y1, x2, y2, px, py)

        return abs(main_area - (area1 + area2 + area3)) < 0.0001

class RasterDrawing:
    def __init__(self, width, height):
        self.canvas = Canvas(width, height)
        self.shapes = []

    def add_shape(self, shape, char='*'):
        self.shapes.append((shape, char))

    def paint(self):
        self.canvas.clear_canvas()
        for shape, char in self.shapes:
            shape.paint(self.canvas, char)

    def display(self):
        self.canvas.display()
```
